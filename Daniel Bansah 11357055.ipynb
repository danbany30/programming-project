```python
import pandas as pd
file_path = "./red_wine.csv"
wineContent  = pd.read_csv(file_path)
```

```python
print("First 10 Rows:")
print(wineContent.head(10))
```

```python
print("Last 20 Rows:")
print(wineContent.tail(20))
```

```python
print(wineContent.info())
```

```python
print(wineContent.describe())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 10))
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 
            'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 
            'pH', 'sulphates', 'alcohol', 'quality']

for i, feature in enumerate(features):
    plt.subplot(4, 3, i+1)
    plt.hist(wineContent[feature].dropna(), bins=20, color='red', alpha=0.8)
    plt.title(feature.replace("_", " ").title())
    plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
```

```python
correlation_matrix = wineContent.corr()
plt.figure(figsize=(10, 8))
plt.title('CORRELATION MATRIX')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.show()
```
